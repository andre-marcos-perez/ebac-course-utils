### **3.3. Listas**

Estrutura de dados do tipo `list` que armazena dados em um formato de lista.

```
salaries = [1500, 3000]

print(salaries)
```

Para acessar um elemento da lista, basta indicar o seu índice, que sempre começa no zero.

```
salary = salaries[0]

print(salary)
```

Para adicionar elementos de uma lista, basta usar o método `append`.

```
salaries.append(5000)

print(salaries)
```

Também é possível acessar mais de um elemento da lista utilizando a técnica de fatiamento.

```
first_two_salaries = salaries[0:2]

print(first_two_salaries)
```

### **3.4. Dicionários**

Estrutura de dados do tipo `dict` que armazena dados em um formato chave/valor.

```
person = {'level': 'Júnior'}

print(person)
```

Os valores de um dicionário podem acessados por suas chaves.

```
level = person['level']

print(level)
```

Para adicionar uma nova chave, basta atribuir um valor.

```
person['salary'] = 1500

print(person)
```

### **3.5. Condições**

Definido pelas palavras reservadas `if` e `else`, é uma estrutura que permite a tomada de decisões lógicas.

```
people = [
  {'level': 'Júnior', 'salary': 1500}, 
  {'level': 'Pleno',  'salary': 3000}, 
  {'level': 'Sênior', 'salary': 5000}
]

junior_salaries = list()
others_salaries = list()
```

```
person = people[0]

if person['level'] == 'Júnior':
  junior_salaries.append(person['salary'])
else:
  others_salaries.append(person['salary'])
```

```
person = people[1]

if person['level'] == 'Júnior':
  junior_salaries.append(person['salary'])
else:
  others_salaries.append(person['salary'])
```

```
person = people[2]

if person['level'] == 'Júnior':
  junior_salaries.append(person['salary'])
else:
  others_salaries.append(person['salary'])
```

```
print(junior_salaries)
print(others_salaries)
```

### **3.6. Repetições**

Definido pela palavra reservada `for`, é uma estrutura que permite a criação de laços de repetição.

```
people = [
  {'level': 'Júnior', 'salary': 1500}, 
  {'level': 'Pleno',  'salary': 3000}, 
  {'level': 'Sênior', 'salary': 5000}
]

junior_salaries = list()
others_salaries = list()
```

```
for person in people:

  if person['level'] == 'Júnior':
    junior_salaries.append(person['salary'])
  else:
    others_salaries.append(person['salary'])
```

```
print(junior_salaries)
print(others_salaries)
```

### **4.1. Salários de nível júnior**

Contar as categorias de salários dos profissionais de dados de nível **júnior** utilizando Python.


 - **Selecionar**

```
import csv

level_index = 17
salary_index = 18
junior_salaries = list()

with open(file='data.csv', mode='r') as fp:
  is_first_line = True
  data = csv.reader(fp, delimiter=',')
  for row in data:
    if is_first_line:
      is_first_line = False
    else:
      level = row[level_index]
      salary = row[salary_index]
      if level == 'Júnior':
        junior_salaries.append(salary)

print(junior_salaries)
```

 - **Contar**

```
unique_junior_salaries_counter = dict()
unique_junior_salaries = list(set(junior_salaries))

for unique_salary in unique_junior_salaries:
  unique_junior_salaries_counter[unique_salary] = junior_salaries.count(unique_salary)

for key, value in unique_junior_salaries_counter.items():
  print(f'{value} - {key}')
```
